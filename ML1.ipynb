```python
df = pd.DataFrame({"A":[1,4,7], "B":[2,5,8], "C":[3,6,9]})
print(df)
```

```python
# Using `iloc[]`
print(df.iloc[0][0])

# Using `loc[]`
print(df.loc[0]['A'])

# Using `at[]`
print(df.at[0,'A'])

# Using `iat[]`
print(df.iat[0,0])
```

```python
new_row = pd.DataFrame([[1, 4, 7]], columns=df.columns)
df = pd.concat([df, new_row], ignore_index=True) # Call pd.concat, assign result back to df
print(df) # Print the updated DataFrame to see the changes
```
