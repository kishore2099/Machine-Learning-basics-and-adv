```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
data = "binary_classification_datasets(1).xlsx"
df= pd.read_excel(data,sheet_name="Medical Diagnosis")
df.head()
```

```python
df.info()
```

```python
missing_values=df.isnull().sum()
missing_values
```

```python
df.describe()
```

```python
percentage_missing=(missing_values/len(df))*100
percentage_missing
```

```python
df=df.fillna(df.median())
df
```

```python
# Normalize numeric features
scaler = StandardScaler()
numeric_features = ['Age', 'Blood Pressure', 'Cholesterol Level', 'BMI']
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
# Plotting histograms for numeric attributes
df[numeric_features].hist(bins=15, figsize=(15, 6), layout=(2, 2))
plt.show()
```

```python
# Plotting box plots for numeric attributes
plt.figure(figsize=(15, 6))
for i, feature in enumerate(numeric_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df[feature])
plt.show()
```

```python
# Creating heatmap for correlations
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Bar plot for target variable distribution
sns.countplot(x='Disease', data=df)
plt.show()
```

```python
# Pie chart for target variable distribution
df['Disease'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.show()
```
